```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ds= pd.read_csv('Mycd.csv')
kfc = pd.read_csv('Mycd.csv')
```

### Getting  % ' Details 'of each posibility 

```python
kfc['inc % in Cases'] = kfc['ActiveCases']/kfc['TotalCases'] * 100
kfc['inc % in Deaths'] = kfc['NewDeaths']/kfc['TotalDeaths'] * 100
kfc['inc % in Recovered'] = kfc['NewRecovered']/kfc['TotalRecovered'] * 100
kfc['AC%'] =  kfc['ActiveCases'] / kfc['Population'] * 100
```

# EDA

```python
kfc.isnull().sum()
```

# Univariate Analysis

#### population Bar by country wise

```python
plt.figure(figsize=(17,5))
plt.bar(kfc['Country'][1:19], kfc['Population'][1:19])
plt.title("population Bar by country wise")
plt.xlabel('Country')
plt.ylabel('population in Million')
plt.show()
```

#### population Bar by continent wise

```python

capc = kfc.groupby(['Continent']).sum()

c = capc.drop('All')


continent = ['Africa' , 'Asia' , 'Austalia/Oceania' , 'Europe',  'North America' , 'South America']

plt.figure(figsize=(10,5))
plt.bar( continent,c['Population'])
plt.title("population Bar by continent wise")
plt.xlabel('Continent')
plt.ylabel('population in Million')
plt.show()
```

#### scattering the Total Cases by Continents

```python
plt.figure(figsize=(10,5))
plt.scatter(continent ,c['TotalCases'],marker = 'H' )
plt.title("scattering the Total Cases by Continents")
plt.legend(["TotalCases"], loc ="upper right")
plt.ylabel(' In Million')
plt.show()
```

#### Plotting Active Case percentage continent Wise

```python

plt.figure(figsize=(10,5))
plt.plot(continent , c['AC%'], 'o:r')
plt.title("Plotting Active Case percentage continent Wise")
plt.legend(["Active Cases percentage "],loc = "upper left")
plt.ylabel(' %')
plt.show()
```

 #### Plotting Serious cases Country Wise

```python

plt.figure(figsize=(19,5))
plt.plot(ds['Country'][1:21] ,ds['Serious ,Critical'][1:21],marker = 'H', ms = 20, mec = 'r')
plt.title("Plotting Active Case percentage country Wise")
plt.legend(["Active Cases percentage "],loc = "upper left")
plt.ylabel('Active case Count')
plt.show()
```

# Bivariate, Multivariate Analysis

#### bar chartting with Country and Active cases showing top 18 countries

```python
plt.figure(figsize=(19,5))
plt.bar(ds['Country'][1:20], ds['ActiveCases'][1:20])
plt.title("bar chartting with Country and Active cases showing top 10 countries")
plt.legend(["Active Cases"],loc = "upper right")
plt.xlabel('Country')
plt.ylabel('population in Million')
plt.show()
```

#### bar chartting with Continent and Active cases 

```python
plt.figure(figsize=(10,5))
plt.bar(continent, c['ActiveCases'])
plt.title("Bar chartting with Continent & Active Cases")
plt.legend(["Active Cases"],loc = "upper left")
plt.ylabel('ActiveCases in Million')
plt.show()
```

#### Plotting the relation b/w Country to Active cases,  Population

```python
plt.figure(figsize=(19,5))
plt.plot(ds['Country'][1:21] ,ds['Population'][1:21], marker = 'o', ms = 10, mec = 'r', mfc = 'r')
plt.plot(ds['ActiveCases'][1:21], marker = '.', ms = 10, mec = '#fffcf2', mfc = '#fffcf2')
plt.title("Plotting the relation b/w Country to Active cases,  Population")
plt.legend(["population", "Active Cases"], loc ="upper right")
plt.ylabel(' In Million')
plt.show()
```

#### Plotting between Active Cases to Total Tests for country

```python
plt.figure(figsize=(19,5))
plt.plot(ds['Country'][1:21] ,ds['TotalCases'][1:21], marker = 'H')
plt.plot(ds['TotalTests'][1:21], marker = 'D', ms = 8, mec = 'y', mfc = 'r' )
plt.title("Plotting the relation b/w Country to TotalCases,  TotalTests")
plt.legend(["TotalCases", "TotalTests"], loc ="upper right")
plt.ylabel(' In Million')
plt.show()
```

#### scattering the relation b/w Continent to Total Cases to New recovered continent Wise

```python
plt.figure(figsize=(10,5))
plt.scatter(continent ,c['TotalCases'],marker = 'H')
plt.scatter(continent ,c['NewRecovered'],marker = 'P')
plt.title("scattering the relation b/w Continent to Total Cases to New recovered")
plt.legend(["TotalCases", "NewRecovered"], loc ="upper right")
plt.ylabel(' In Million')
plt.show()
```

#### scattering  inc % in Cases and inc % in Recovery continent Wise

```python
plt.figure(figsize=(10,5))
plt.scatter(continent ,c['inc % in Cases'], marker = 'o')
plt.scatter(continent ,c['inc % in Recovered'], marker = 's')
plt.title("scattering  inc % in Cases and inc % in Recovery")
plt.legend(["inc % in Cases", "inc % in Recovered"], loc ="upper right")
plt.ylabel(' %')
plt.show()
```
