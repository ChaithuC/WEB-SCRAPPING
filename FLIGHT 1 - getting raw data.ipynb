## getting the URL and URL data and getting the headers of the table

```python

url = 'https://www.worldometers.info/coronavirus/#countries'
mypage = requests.get(url)
soup = BeautifulSoup(mypage.text, 'lxml')

mytable = soup.find('table', {'id': 'main_table_countries_today'})

headers = []

for i in mytable.find_all('th'):
    title = i.text
    headers.append(title)
```


## preparing a  data frame and getting the whole raw data from the website 


```python
 
df = pd.DataFrame(columns = headers)


for row in mytable.find_all('tr')[1:]:
    data = row.find_all('td')
    row_data = [td.text.strip() for td in data]
    length = len(df)
    df.loc[length] = row_data
```
