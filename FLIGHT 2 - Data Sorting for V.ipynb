## Importings

## Reading CSV

```python
dff = pd.read_csv('rawdata.csv')
```

## Filtering 

### Preparing Data to convert into int

```python
dff['TotalCases']=dff['TotalCases'].str.replace(',','')
dff['NewCases']=dff['NewCases'].str.replace(',','')
dff['NewCases']=dff['NewCases'].str.replace('+','')


dff['TotalDeaths']=dff['TotalDeaths'].str.replace(',','')

dff['NewDeaths']=str(dff['NewDeaths'])

dff['NewDeaths']=dff['NewDeaths'].str.replace(',','')
dff['NewDeaths']=dff['NewDeaths'].str.replace('+','')

dff['TotalRecovered']=dff['TotalRecovered'].str.replace(',','')
dff['NewRecovered']=dff['NewRecovered'].str.replace(',','')
dff['NewRecovered']=dff['NewRecovered'].str.replace('+','')
dff['Serious,Critical']=dff['Serious,Critical'].str.replace(',','')


dff['ActiveCases']=dff['ActiveCases'].str.replace(',','')
dff['TotalTests']=dff['TotalTests'].str.replace(',','')
dff['Population']=dff['Population'].str.replace(',','')
```

```python
### Removing 'nan' and 'NaN' cells
```

```python
c= dff
```

```python
Continent = c['Continent']
Country = c['Country,Other']
Population=[]
TotalCases= []
NewCases = []
TotalDeaths= []
NewDeaths= []
TotalRecovered = []
NewRecovered = []
ActiveCases = []
Serious = []
TotalTests = []



#Population
for i in c['Population']:
    if str(i)=='nan':
        Population.append(0.0)
    else:
        Population.append(i)
    
    
    
# TotalCases
for i in c['TotalCases']:
    if str(i)=='nan':
        TotalCases.append(0.0)
    else:
        TotalCases.append(i)
        
        
#NewCases
for i in c['NewCases']:
    if str(i)=='nan':
        NewCases.append(0.0)
    else:
        NewCases.append(i)
   
        
        
# TotalDeaths
for i in c['TotalDeaths']:
    if str(i)=='nan':
        TotalDeaths.append(0.0)
    else:
        TotalDeaths.append(i)
        
        
# NewDeaths
        
for i in c['NewDeaths']:
    if str(i)=='nan'or 'NaN':
        NewDeaths.append(0.0)
    else:
        NewDeaths.append(i)
    
    
    
# TotalRecovered
for i in c['TotalRecovered']:
    if str(i)=='nan':
        TotalRecovered.append(0.0)
    else:
        TotalRecovered.append(i)
        
        
# NewRecovered
for i in c['NewRecovered']:
    if str(i)=='nan':
        NewRecovered.append(0.0)
    else:
        NewRecovered.append(i)
    
    


# ActiveCases
for i in c['ActiveCases']:
    if str(i)=='nan':
        ActiveCases.append(0.0)
    else:
        ActiveCases.append(i)
    
    
# Serious,Critical
for i in c['Serious,Critical']:
    if str(i)=='nan':
        Serious.append(0.0)
    else:
        Serious.append(i)
    
#TotalTests
for i in c['TotalTests']:
    if str(i)=='nan':
        TotalTests.append(0.0)
    else:
        TotalTests.append(i)
   
```

### Making to Series


```python
Population =pd.Series(Population)

TotalCases =pd.Series(TotalCases)

TotalDeaths =pd.Series(TotalDeaths)

NewDeaths =pd.Series(NewDeaths)

TotalRecovered =pd.Series(TotalRecovered)

NewRecovered =pd.Series(NewRecovered)

NewCases  =pd.Series(NewCases )


ActiveCases  =pd.Series(ActiveCases )

Serious  =pd.Series(Serious )
TotalTests  =pd.Series(TotalTests )
```

### converting to numeric to do operations

```python

Population =pd.to_numeric(Population)

TotalCases =pd.to_numeric(TotalCases)

TotalDeaths =pd.to_numeric(TotalDeaths)

NewDeaths =pd.to_numeric(NewDeaths)

TotalRecovered =pd.to_numeric(TotalRecovered)

NewRecovered =pd.to_numeric(NewRecovered)

NewCases=pd.to_numeric(NewCases)


ActiveCases=pd.to_numeric(ActiveCases)


Serious=pd.to_numeric(Serious)

TotalTests=pd.to_numeric(TotalTests)
```

```python

### Converting to DATA FRAME
###  Adding Continent ,Country
```

```python
newframe =pd.DataFrame({ 'Continent': Continent, 'Country': Country ,'Population': Population,  'TotalCases':TotalCases ,'TotalDeaths': TotalDeaths,'NewDeaths': NewDeaths, 'TotalRecovered': TotalRecovered,'NewRecovered': NewRecovered,'NewCases': NewCases,'ActiveCases': ActiveCases,'Serious ,Critical': Serious, 'TotalTests': TotalTests  })

 
newframe['Population'] = newframe['Population'].astype(int)

newframe.to_csv('Mycd.csv')
```
